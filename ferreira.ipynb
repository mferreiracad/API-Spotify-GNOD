```python
def trans_features(X,y):
    from sklearn.cluster import KMeans
    X = X.drop(['Song Uri','Song','Artist Uri','Artist','cluster'],axis = 1)
    scaler = StandardScaler()
    X_prep = scaler.fit_transform(X)
    y = scaler.transform(y)
    kmeans = KMeans(n_clusters=22, random_state=13)
    kmeans.fit(X_prep)
    c_prediction = kmeans.predict(y)
    return c_prediction
```

```python
def features_ferr(z):
    danceability_ = []
    energy_ = []
    key_ = []
    loudness_ = []
    mode_ = []
    speechiness_ =[]
    acousticness_ = []
    instrumentalness_ = []
    liveness_ = []
    valence_ = []
    tempo_ = []
    time_signature_ = []
    duration_ms_ = []
    
    x = z[0]
    try:
        danceability_.append(x['danceability'])
    except:
        danceability_.append(1000)
    try:
        energy_.append(x['energy'])
    except:
        energy_.append(1000)
    try:    
        key_.append(x['key'])
    except:
        key_.append(1000)
    try:    
        loudness_.append(x['loudness'])
    except:
        loudness_.append(1000)
    try:    
        mode_.append(x['mode'])
    except:
        mode_.append(1000)
    try:    
        speechiness_.append(x['speechiness'])
    except:
        speechiness_.append(1000)
    try:    
        acousticness_.append(x['acousticness'])
    except:
        acousticness_.append(1000)
    try:    
        instrumentalness_.append(x['instrumentalness'])
    except:
        instrumentalness_.append(1000)
    try:    
        liveness_.append(x['liveness'])
    except:
        liveness_.append(1000)
    try:    
        valence_.append(x['valence'])
    except:
        valence_.append(1000)
    try:    
        tempo_.append(x['tempo'])
    except:
        tempo_.append(1000)
    try:    
        time_signature_.append(x['time_signature'])
    except:
        time_signature_.append(1000)
    try:    
        duration_ms_.append(x['duration_ms'])
    except:
        duration_ms_.append(1000)
    
    song_feat = pd.DataFrame({"Danceability":danceability_, "Energy":energy_,"Key":key_, \
                              "Loudness":loudness_,"Mode":mode_, "Speechiness": speechiness_, \
                              "Acousticness":acousticness_, "Instrumentalness":instrumentalness_,\
                              "Liveness":liveness_, "Valence": valence_, "Tempo":tempo_, \
                              "Time_signature":time_signature_, "Duration_ms":duration_ms_})
    return song_feat
```
